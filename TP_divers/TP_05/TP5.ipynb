```python
#Suppresion des valeurs vides dans la colonne id_web, car sans ces valeurs, impossible de faire une liaison entre les deux fichiers

liasion_clean_df = liaison_df.dropna()
liasion_clean_df
#91 product_id ont une valeur id_web qui n'existe pas, il reste 734 produits avec une id_web
```

```python
web_df
```

```python
#On renome la colonne sku en product_id
web_df.rename(columns={'sku': 'id_web'}, inplace=True)
web_df
```

```python
#Suppresion des colonnes vides ou avec la meme valeur

web_df = web_df.drop(["virtual"], axis=1)
web_df = web_df.drop(["downloadable"], axis=1)
web_df = web_df.drop(["rating_count"], axis=1)
web_df = web_df.drop(["tax_class"], axis=1)
web_df = web_df.drop(["post_password"], axis=1)
web_df = web_df.drop(["post_content_filtered"], axis=1)
web_df
```

```python
#Suppresion des colonnes sans product_id
web_df = web_df.dropna(subset=['id_web'])
web_df
```

```python
#Suppresion des valeurs attachment dans la colonne post_type, 
#elles correspondent seulement aux images du site web de chaque bouteille de vin, 
#elles n'ont aucune valeurs analytique 

web_df = web_df[web_df["post_type"]!="attachment"]
web_df
```

```python
#Après la mise en place des filtres précedents, 
#on peut supprimer d'autres colonnes qui n'ont plus de valeurs analytiques

web_df = web_df.drop(["average_rating"], axis=1)

web_df = web_df.drop(["tax_status"], axis=1)
web_df = web_df.drop(["post_content"], axis=1)
web_df = web_df.drop(["post_parent"], axis=1)
web_df = web_df.drop(["menu_order"], axis=1)
web_df = web_df.drop(["post_mime_type"], axis=1)
web_df = web_df.drop(["comment_count"], axis=1)

web_df
```

```python
# 2 Preparation des tables pour la fusion en une table unique

#On constate que le nombre de produit n'est pas le meme entre la table erp et la table web.

print(len(list(set(erp_df["product_id"]))))
print(len(list(set(liasion_clean_df["product_id"]))))
print(len(list(set(liasion_clean_df["id_web"]))))
print(len(list(set(web_df["id_web"]))))
```

```python
#Tous les product_id de la table liasion_clean_df ne sont dans la table web_df
print(len(set(liasion_clean_df["id_web"]).difference(set(web_df["id_web"]))))

#Mais tous les product_id de la table erp_df sont pas dans la table liasion_clean_df
print(len(set(web_df["id_web"]).difference(set(liasion_clean_df["id_web"]))))
```

```python
#On stocke les id_web sans qui ne sont pas dans la table liaison_clean_df dans la table liaison_clean_df_orphelin
liaison_clean_df_orphelin = liasion_clean_df[-liasion_clean_df["id_web"].isin(web_df["id_web"])]

liaison_clean_df_orphelin
```

```python
#On garde dans la table liasion_clean_df uniquement les id_web qui sont dans la table liaison_clean_df
liasion_clean_df = liasion_clean_df[liasion_clean_df["id_web"].isin(web_df["id_web"])]

liasion_clean_df
```

```python
#Tous les product_id de la table liasion_clean_df sont dans la table erp_df
print(len(set(liasion_clean_df["product_id"]).difference(set(erp_df["product_id"]))))

#Mais tous les product_id de la table erp_df ne sont pas dans la table liasion_clean_df
print(len(set(erp_df["product_id"]).difference(set(liasion_clean_df["product_id"]))))
```

```python
#On stocke les products_id sans id_web dans la table erp_orphelin_df
erp_orphelin_df = erp_df[-erp_df["product_id"].isin(liasion_clean_df["product_id"])]

erp_orphelin_df
```

```python
#On supprimer les products_id de la table erp_df pour pouvoir ensuite fusionner les données sans risque

erp_df = erp_df[erp_df["product_id"].isin(liasion_clean_df["product_id"])]

erp_df
```

```python
#Toutes nos colonnes de fusion sont de meme longueur, on peux maintenant fusionner les données
print(len(list(set(erp_df["product_id"]))))
print(len(list(set(liasion_clean_df["product_id"]))))
print(len(list(set(liasion_clean_df["id_web"]))))
print(len(list(set(web_df["id_web"]))))
```

```python
data_fusionned = pd.merge(erp_df, liasion_clean_df,on=["product_id"])
data_fusionned
```

```python
data_fusionned = pd.merge(data_fusionned, web_df,on=["id_web"])
data_fusionned
```

```python
#3 analyse des données

#Calcul du chiffre d’affaires par produit

data_fusionned["Chiffre d'affaire"] = data_fusionned["price"] * data_fusionned["total_sales"]

#on trie les produis selon le chiffre d'affaires
data_fusionned = data_fusionned.sort_values(["Chiffre d'affaire"], ascending=False)

data_fusionned
```
